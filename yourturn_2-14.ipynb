```python
s = {'David': 3.2, 'Ray': 6,  'Zac': 12, 'Sam': 9.5}
total = 0
f = 'Unknown'
for x in s:
    total += s[x]
    if f == 'Unknown' or s[x] > s[f]:
        f = x
print(f'total {total} KM, and {f} run farest')
```

```python
s = {'David': 3.2, 'Ray': 6,  'Zac': 12, 'Sam': 9.5}
f = max(s.items(), key=lambda x: x[1])[0]
print(f'total {sum(s.values())} KM, and {f} run farest')
```
