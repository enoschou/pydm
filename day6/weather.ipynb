```python
import requests
from bs4 import BeautifulSoup

def lookup(site):
    # look up site id
    url = 'https://www.cwa.gov.tw/Data/js/Observe/OSM/C/STMap.json'
    r = requests.get(url)
    for d in r.json():
        if d['STname'] == site:
            return d['ID']

def grab(site):
    siteid = lookup(site)
    if not siteid:
        return None
    
    # get page content
    url = f'https://www.cwa.gov.tw/V8/C/W/Observe/MOD/24hr/{siteid}.html'
    r = requests.get(url)
    
    # parsing
    soup = BeautifulSoup(r.text, 'html.parser')
    a = soup.find('tr')
    o = a.find(headers="time").text
    r = float(a.find(headers="rain").text)
    h = float(a.find(headers="hum").text)/100
    t = float(a.find(class_="tem-C is-active").text)

    # {'O': '04/24 13:00', 'T': 29.1, 'H': 0.67, 'R': 0.5} 
    return {'O': o, 'T': t, 'H': h, 'R': r}
    
# output
site = input('請輸入測站名稱：　')
w = grab(site)
if w:
    print(f'{w["O"]} {w["T"]:.1f}度, {w["H"]}, {w["R"]:.1f}mm')
else:
    print('no such site')
```

```python
import requests
from bs4 import BeautifulSoup

class WeaG:
    
    url = 'https://www.cwa.gov.tw/Data/js/Observe/OSM/C/STMap.json'
    
    def __init__(self):
        r = requests.get(__class__.url)
        self.sitemap = {d['STname']: d['ID'] for d in r.json()}
        r.close()

    def lookup(self, site):
        return self.sitemap.get(site)

    def grab(self, site):
        siteid = self.lookup(site)
        if not siteid:
            return None
        
        # get page content
        url = f'https://www.cwa.gov.tw/V8/C/W/Observe/MOD/24hr/{siteid}.html'
        r = requests.get(url)
        
        # parsing
        soup = BeautifulSoup(r.text, 'html.parser')
        a = soup.find('tr')
        o = a.find(headers="time").text
        r = float(a.find(headers="rain").text)
        h = float(a.find(headers="hum").text)/100
        t = float(a.find(class_="tem-C is-active").text)
    
        # {'O': '04/24 13:00', 'T': 29.1, 'H': 0.67, 'R': 0.5} 
        return {'O': o, 'T': t, 'H': h, 'R': r}


w = WeaG()
print(w.grab('臺北'))
# # {'O': '04/24 13:00', 'T': 29.1, 'H': 0.67, 'R': 0.5}
# print(w.grab('新竹'))
# # {'O': '04/24 13:00', 'T': 29.1, 'H': 0.67, 'R': 0.5} 
print(w.grab('aaaa'))
# # None
# print(w.grab('七美'))
# {'O': '04/24 13:00', 'T': 29.1, 'H': 0.67, 'R': 0.5} 
```
