# Solutions of Missing Number

```python
nums = [3,0,1]
nums = [0,1]
nums = [9,6,4,2,3,5,7,0,1]

Solution().missingNumber(nums)
```

```python
# solution 1

class Solution:
    def missingNumber(self, nums):
        pool = set(range(len(nums)+1))
        for n in nums:
            pool.remove(n)
        return pool.pop()
```

```python
# solution 2

class Solution:
    def missingNumber(self, nums):
        len_nums = len(nums) + 1
        flags = [False] * len_nums
        for n in nums:
            flags[n] = True
        for i in range(len_nums):
            if not flags[i]:
                return i
```

```python
# solution 3

class Solution:
    def missingNumber(self, nums):
        flag = 0
        for n in nums:
            flag |= 1 << n
        t, i = 1, 0
        while flag & t:
            t <<= 1
            i += 1
        return i
```

```python
# solution 4

class Solution:
    def missingNumber(self, nums):
        return sum(range(len(nums)+1)) - sum(nums)
```

```python
# solution 5

class Solution:
    def missingNumber(self, nums):
        s = len_nums = len(nums)
        for i in range(len_nums):
            s += i - nums[i]
        return s
```

```python
# solution 6

class Solution:
    def missingNumber(self, nums):
        s = 0
        for i, v in enumerate(nums):
            s += i - v
        return s + i + 1
```
