## Solution

```python
from random import sample

def guess(game):
    s = input(f'round [{i}], please guess: ')
    s = ''.join(s.split())
    if len(s) == len(game) and (len(set(s)) == len(s)) and s.isdigit():
        return [int(i) for i in s]

def judge(game, guess):
    a, b = 0, 0
    for i in range(len(game)):
        if game[i] == guess[i]:
            a += 1
        elif game[i] in guess:
            b += 1
    return a, b


game = sample(range(10), 4)
print(game)

for i in range(10):
    # user guess and validate
    g = guess(game)
    if g == None:
        print('bad input, again')
        continue
    
    # judge ?A ?B
    a, b = judge(game, g)
    
    # output result
    if a == 4:
        print('you win')
        break
    else:
        print(f'{a}A, {b}B')
```
