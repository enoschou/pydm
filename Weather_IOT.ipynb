```python
import requests

url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/O-A0003-001?Authorization='
r = requests.get(url)

print(r)
print(r.status_code)
```

```python
print(r.text)
```

```python
j = r.json()
print(j)
```

```python
j
```

```python
locs = j['records']['location']
locs
```

```python
print(f'{len(locs)} sites')
```

```python
for loc in locs:
    if loc['locationName'] == '臺北':
        print(loc)
```
