```python
from flask import Flask, request

from yourturn_5_2 import WeaG

w = WeaG('')
app = Flask(__name__)

@app.route('/')
def wea():
    loc = request.args.get('location')
    
    if loc:
        content = f"{w.grab(loc)}"
    else:
        content = '''
        <form>
            <input type="text" name="location"> <input type="submit" value="確定">
        </form>
        '''
    
    return content
```

```python
app.run()
```
