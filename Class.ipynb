```python
import os
from glob import glob

class Trees1:
    _path = None
    
    def __init__(self, path):
        if self._isvalid(path):
            self._path = path

    def count(self, dataset, sp):
        if not (self._isvalid(dataset) and self._isvalid(sp)):
            return -1
        w = os.path.join(self._path, dataset, sp, '*.jpg')
        count = len(glob(w))
        
        return count
    
    def _isvalid(self, s):
        return bool(s) and 2 <= len(s) <= 5
    
t1 = Trees1('trees')
print(f"train DR: {t1.count('training', 'DR')}")
print(f"test AS: {t1.count('test', 'AS')}")
```

```python
import os
from glob import glob

class Trees2:
    @staticmethod
    def count(path, dataset, sp):
        w = os.path.join(path, dataset, sp, '*.jpg')
        count = len(glob(w))
        
        return count

print(f"train DR: {Trees2.count('trees', 'train', 'DR')}")
print(f"test AS: {Trees2.count('trees', 'test', 'AS')}")
```

```python
import os
from glob import glob

class Trees3:
    _path = None
    
    @classmethod
    def __init__(cls, path):
        cls._path = path

    @classmethod
    def count(cls, dataset, sp):
        w = os.path.join(cls._path, dataset, sp, '*.jpg')
        count = len(glob(w))
        
        return count
    
t3 = Trees3('trees')
print(f"train DR: {t3.count('train', 'DR')}")
print(f"test AS: {Trees3.count('test', 'AS')}")
```
