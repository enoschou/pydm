```python
# full 剪刀石頭布
from random import choice

allow = ['剪刀', '石頭', '布']
computer = choice(allow)
user = input(f'{', '.join(allow)}:')
if user in allow:
    print(f'computer: {computer} vs user: {user}')
    
    if ((user == '石頭' and computer == '剪刀') or 
        (user == '布' and computer == '石頭') or 
        (user == '剪刀' and computer == '布')):  # user win
        print('user win')
    elif user == computer:  # tie
        print('tie')
    else:  # user lose
        print('user lose')
else:
    print('輸入錯誤')
```

```python
s = {'David': 3.2, 'Ray': 6, 'Bill': 12, 'Sam': 9.5}
_max = -1
_name = ''
for name in s:
    print(name, s[name])
    if s[name] > _max:
        _max = s[name]
        _name = name

print(f'farest is {_name}', _max)
```

```python
s = {'David': 3.2, 'Ray': 6, 'Bill': 12, 'Sam': 9.5}
ss = list(s.items())
print(ss)

if ss:
    _name, _max = ss[0]
    for n, km in ss[1:]:
        if km > _max:
            _max = km
            _name = n
    print(f'farest is {_name}', _max)
```

```python
s = [5, 2, 7, 1, 3]

print(s)

for j in range(len(s)-1):
    for i in range(len(s)-1-j):   # single round
        print(f's[{i}]={s[i]}, s[{i+1}]={s[i+1]}')
        if s[i] > s[i+1]:
            s[i], s[i+1] = s[i+1], s[i]
        #print(s)
    print(j, s)
```
