```python
"""
original data
"""
d = {
    'David': {'h':180, 'w':85},
    'John': {'h':172, 'w':72},
    'Mary': {'h':158, 'w':51},
    'Lora': {'h':160, 'w':49},
    'Bill': {'h':168, 'w':70},
    'Elsa': {'h':154, 'w':55},
    'Golden': {'h':176, 'w':69},
    'Jason': {'h':182, 'w':77},
    'Larry': {'h':160, 'w':64},
    'Michelle': {'h':155, 'w':57}
}
```

### a. 計算每個人的 BMI 並儲存至上述 dict

```python
"""
Create BMI data
"""
for name in d:
    bmi = d[name]['w'] / (d[name]['h'] / 100) ** 2  # bmi algorithm: weight(kg)/height(m)**2
    d[name]['bmi'] = bmi

from pprint import pprint
pprint(d)
```

### b. 運用上述 dict 計算並印出所有人平均 BMI、BMI 變異數、BMI 標準差

```python
# Mean BMI

# bmis = []
# for e in d.values():
#     bmis.append(e['bmi'])

# bmis = [e['bmi'] for e in d.values()]
# mean_bmi = sum(bmis) / len(bmis)

mean_bmi = sum(e['bmi'] for e in d.values()) / len(d)

print(mean_bmi)
```

```python
# Variance BMI

# bmis_diff_square = []
# for e in d.values():
#     bmis_diff_square.append((e['bmi']-mean_bmi)**2)
# var_bmi = sum(bmis_diff_square) / len(bmis_diff_square)

# bmis_diff_square = [(e['bmi']-mean_bmi)**2 for e in d.values()]
# var_bmi = sum(bmis_diff_square) / len(bmis_diff_square)

var_bmi = sum((e['bmi']-mean_bmi)**2 for e in d.values()) / len(d)

print(var_bmi)
```

```python
# Standard Deviation BMI
std_bmi = var_bmi ** 0.5

print(std_bmi)
```

```python
print(f'mean bmi is {mean_bmi:.3f} kg/m\xB2')
print(f'var bmi is {var_bmi:.3f} (kg/m\xB2)\xB2')
print(f'std bmi is {std_bmi:.3f} kg/m\xB2')
```

### c. 找出 BMI 最小的三位學生

```python
# brute-force solution

n = len(d)
smallest = min(3, n)  # prevent n is smaller than smallest

# create (bmi, name) pairs
bmi_names = [(d[name]['bmi'], name) for name in d]

# limited selection sort
for j in range(smallest):
    for i in range(j+1, n):
        if bmi_names[i][0] < bmi_names[j][0]:
            bmi_names[i], bmi_names[j] = bmi_names[j], bmi_names[i]

print(', '.join(name for _, name in bmi_names[:smallest]))
```

```python
# bubble solution

n = len(d)
smallest = min(3, n)  # prevent n is smaller than smallest

# create (bmi, name) pairs
bmi_names = [(d[name]['bmi'], name) for name in d]

# bubble sort according to bmi inside pair
for j in range(n-1):
    for i in range(n-1-j):
        if bmi_names[i][0] > bmi_names[i+1][0]:
            bmi_names[i], bmi_names[i+1] = bmi_names[i+1], bmi_names[i]

print(', '.join(name for _, name in bmi_names[:smallest]))
```

```python
# cost-down bubble solution

n = len(d)
smallest = min(3, n)  # prevent n is smaller than smallest

# create (bmi, name) pairs
bmi_names = [(d[name]['bmi'], name) for name in d]

# bubble sort according to bmi inside pair, but not fully sorted
for j in range(min(n-1, smallest)):
    for i in range(n-1-j):
        if bmi_names[i][0] < bmi_names[i+1][0]:  # < for bubble up smallest
            bmi_names[i], bmi_names[i+1] = bmi_names[i+1], bmi_names[i]

print(', '.join(name for _, name in bmi_names[-1:-1-smallest:-1]))
```

```python
# simple solution, leverage sorted

n = len(d)
smallest = min(3, n)  # prevent n is smaller than smallest

# create (bmi, name) pairs
bmi_names = [(d[name]['bmi'], name) for name in d]

# sort via sorted
print(', '.join(name for _, name in sorted(bmi_names)[:smallest]))
```
