# One-Line Statements

## One-line if

```python
# 檢查 x 是否小於 y
x = 60
y = 100
if x < y:
    print('x is less than y')
```

```python
# 檢查 x 是否小於 y
x = 60
y = 100
if x < y: print('x is less than y')
```

## One-line if-else

```python
# 計算 x 與 y 的差 (絕對值)
x = 60
y = 100
if x > y:
    result = x - y
else:
    result = y - x
print(result)
```

```python
# 計算 x 與 y 的差 (絕對值)
x = 60
y = 100
result = x - y if x > y else y - x
print(result)
```

```python
# 計算 x 與 y 的差 (絕對值)
x = 60
y = 100
result = x > y and x - y or y - x
print(result)
```

## One-line for loop

```python
# 1 ~ 15 int list
y = []
for i in range(1, 16):
    y.append(i)
print(y)

# 1 ~ 15 odd int list
q = []
for i in range(1, 16):
    if i % 2 == 1:
        q.append(i)
print(q)
```

```python
# 1 ~ 15 int list
yy = [i for i in range(1, 16)]
print(yy)

# 1 ~ 15 odd int list
qq = [i for i in range(1, 16) if i % 2 == 1]
print(qq)
```
