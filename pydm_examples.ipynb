```python
import os
from glob import glob

class Trees1:
    _path = None
    
    def __init__(self, path):
        if self._isvalid(path):
            self._path = path

    def count(self, dataset, sp):
        if not (self._isvalid(dataset) and self._isvalid(sp)):
            return -1
        w = os.path.join(self._path, dataset, sp, '*.jpg')
        count = len(glob(w))
        
        return count
    
    def _isvalid(self, s):
        return bool(s) and 2 <= len(s) <= 5
    
t1 = Trees1('trees')
print(f"train DR: {t1.count('training', 'DR')}")
print(f"test AS: {t1.count('test', 'AS')}")
```

```python
import random

class Dealer:
    
    def __init__(self):
        self._cards = []
        for i in ['SPADE', 'HEART', 'DIAMOND', 'CLUB']:
            for j in range(1, 14):
                self._cards.append((i, j))
    
    def deal(self):
        if self._cards:
            x = random.choice(self._cards)
            self._cards.remove(x)
            return x
        else:
            return None
```

```python
d = Dealer()

for c in range(53):
    print(d.deal())
```

```python
import random

class Dealer:
    
    def __init__(self):
        self._cards = []
        self.reset()
    
    def deal(self):
        if self._cards:
            x = random.choice(self._cards)
            self._cards.remove(x)
            return x
        else:
            return None
    
    def reset(self):
        self._cards.clear()
        for i in ['SPADE', 'HEART', 'DIAMOND', 'CLUB']:
            for j in range(1, 14):
                self._cards.append((i, j))
```

```python
d = Dealer()

for c in range(53):
    print(d.deal())

d.reset()
for c in range(53):
    print(d.deal())
```

### Packing Arbitrary Keyword Arguments

```python
def update1(user, h=None, w=None, g=None):
    if user in d:
        if h != None:
            d[user]['h'] = h
        if w != None:
            d[user]['w'] = w
        if g != None:
            d[user]['g'] = g
```

```python
name = 'David'
print(f'{name} before: {d[name]}')
update1(user=name, h=175)
print(f'{name} after:  {d[name]}')

name = 'Mary'
print(f'{name} before: {d[name]}')
update1(user='Mary', w=60)
print(f'{name} after:  {d[name]}')

name = 'Bill'
print(f'{name} before: {d[name]}')
update1(user=name, h=173, w=77)
print(f'{name} after:  {d[name]}')
```

```python
def update2(user, **kwargs):
    if user in d:
        for a in kwargs:
            if a in d[user]:
                d[user][a] = kwargs[a]
```

```python
name = 'David'
print(f'{name} before: {d[name]}')
update2(user=name, h=175)
print(f'{name} after:  {d[name]}')

name = 'Mary'
print(f'{name} before: {d[name]}')
update2(user='Mary', w=60)
print(f'{name} after:  {d[name]}')

name = 'Bill'
print(f'{name} before: {d[name]}')
update2(user=name, h=173, w=77)
print(f'{name} after:  {d[name]}')
```

## <div id="pythread"><font color="blue">Python Threading</font></div>

```python
import threading
from time import sleep

finished = False

def show(secs):
    i = 0
    while not finished:
        print(i*secs)
        sleep(secs)
        i += 1

def timer(secs):
    global finished
    
    print(f'countdown for {secs} ses.')
    finished = False
    sleep(secs)
    finished = True
    print('timeout')
    
t = threading.Thread(target=timer, args=(5,))
t.daemon = True
t.start()
show(1.3)
t.join()
print('finished')
```

## <div id="pythread"><font color="blue">Python Iterator</font></div>

```python
class Iter:

    def __init__(self):
        pass

    def __iter__(self):  # reset iterator
        return self  # always return self

    def __next__(self):  # get item
        raise StopIteration  # return item or stop iteration

    def __len__(self):  # get length of iterable
        pass  # return length of iterable

    def __getitem__(self, key):  # get item of specific subscription
        pass  # return subscripted item
```

```python
class Alphabetiter:

    def __init__(self):
        self.alphabets = 'abcdefghijklmnopqrstuvwxyz'
        self.i = -1

    def __iter__(self):
        self.i = -1
        return self

    def __next__(self):
        self.i += 1
        if self.i < len(self.alphabets): return self.alphabets[self.i]
        raise StopIteration

    def __len__(self):
        return len(self.alphabets)

    def __getitem__(self, key):
        return self.alphabets[key]
```

```python
a = Alphabetiter()
for x in a:
    print(x, end='')
print()
for i in range(len(a)):
    print(a[i], end='')
```
