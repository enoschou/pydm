## 完整版猜拳，讓用戶輸入剪刀、石頭、布，與電腦比輸贏

```python
# solution 1, basic logic

from random import choice

allow = ['剪刀', '石頭', '布']
computer = choice(allow)
you = input(allow)
if you not in allow:
    print('bad input')
else:
    if you == computer:
        result = 'tie'
    elif you == '剪刀' and computer == '布' or you == '石頭' and computer == '剪刀' or you == '布' and computer == '石頭':
        result = 'you win'
    else:
        result = 'you lose'
    print(f'you {you} vs computer {computer}: {result}')
```

```python
# solution 2, computed logic

from random import choice

allow = ['剪刀', '石頭', '布']
computer = choice(allow)
you = input(allow)
if you not in allow:
    print('bad input')
else:
    d = allow.index(you) - allow.index(computer)
    if not d:
        result = 'tie'
    elif d in (1, -2):
        result = 'you win'
    else:
        result = 'you lose'
    print(f'you {you} vs computer {computer}: {result}')
```
