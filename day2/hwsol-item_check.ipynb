## Tasks
承 Homework Composite Data Structure，以 dict 建立 10 位學生資料<br>
a. 隨意輸入目標姓名，印出身高資訊，符合下列規格<br>

<ol>
    <li>輸入姓名 "Chloe"，印出 "no Chloe"</li>
    <li>輸入姓名 "Jason"，印出 "Jason's height is 182cm"</li>
</ol>
b. 同 a，但不使用 membership operator 完成相同任務

## Data
<br>
<tt>
David&nbsp;&nbsp;&nbsp;&nbsp;180cm 85kg<br>
John&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;172cm 72kg<br>
Mary&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;158cm 51kg<br>
Lora&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;160cm 49kg<br>
Bill&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;168cm 70kg<br>
Elsa&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;154cm 55kg<br>
Golden&nbsp;&nbsp;&nbsp;176cm 69kg<br>
Jason&nbsp;&nbsp;&nbsp;&nbsp;182cm 77kg<br>
Larry&nbsp;&nbsp;&nbsp;&nbsp;160cm 64kg<br>
Michelle 155cm 57kg<br>
</tt>

```python
"""
original data
"""
d = {
    'David': {'h':180, 'w':85},
    'John': {'h':172, 'w':72},
    'Mary': {'h':158, 'w':51},
    'Lora': {'h':160, 'w':49},
    'Bill': {'h':168, 'w':70},
    'Elsa': {'h':154, 'w':55},
    'Golden': {'h':176, 'w':69},
    'Jason': {'h':182, 'w':77},
    'Larry': {'h':160, 'w':64},
    'Michelle': {'h':155, 'w':57}
}
```

### a. 隨意輸入目標姓名，印出身高資訊

```python
target = 'Chloe'

if target in d:
    print(f'{target}\'s height is {d[target]['h']}cm')
else:
    print(f'no {target}')

target = 'Jason'

if target in d:
    print(f'{target}\'s height is {d[target]['h']}cm')
else:
    print(f'no {target}')
```

### b. 不使用 membership operator 完成相同任務

```python
target = 'Chloe'

height = d.get(target, {}).get('h')
if height:
    print(f'{target}\'s height is {height}cm')
else:
    print(f'no {target}')

target = 'Jason'

height = d.get(target, {}).get('h')
if height:
    print(f'{target}\'s height is {height}cm')
else:
    print(f'no {target}')
```
