```python
s = ['T-ara', 'BLACKPINK', 'Wonder Girls', 'TWICE']
f = [7, 4, 5, 9]
```

```python
i = 0
while i < len(s):
    if f[i] == 5:
        print(f'{s[i]} is 5')
    i += 1
```

```python
i = 0
while i < len(s):
    print(f'run {i}')
    if f[i] == 5:
        print(f'{s[i]} is 5')
    i += 1
```

```python
i = 0
flag = False
while flag == False and i < len(s):
    print(f'run {i}')
    if f[i] == 5:
        print(f'{s[i]} is 5')
        flag = True
    i += 1
```

```python
i = 0
found = False
while not found and i < len(s):
    print(f'run {i}')
    if f[i] == 5:
        print(f'{s[i]} is 5')
        found = True
    i += 1
```
