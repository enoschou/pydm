```python
'''method 1'''

import random
import time

c = 10000000

start = time.time()
# x = [random.randint(0, 1000) for i in range(c)]
x = []
for i in range(c):
    x.append(random.randint(0, 1000))
end = time.time()
print(f'method 1: by randint - {end-start:.3f} secs')
```

```python
'''method 2'''

import random
import time

c = 10000000

start = time.time()
x = random.choices(range(0, 1000), k=c)
end = time.time()
print(f'method 2: by choices - {end-start:.3f} secs')
```
