## Tasks
製作猜數字遊戲，由電腦隨機出 4 個不同數字讓用戶猜，每次用戶猜測的 4 個數字會由電腦輸出 `?A?B` 的結果；每個遊戲在用戶完全猜對 (4A0B) 或猜測 10 次之後結束。

````python
import random


def _extract_guess(guess: str, digits: int) -> list:
    g = ''.join(guess.split())
    if not g.isdigit() or len(set(g)) != digits:
        return []
    return [int(n) for n in g]

def _judge(guess: list[int], game: list[int]) -> tuple[int, int]:
    a = b = 0
    for gu, ga in zip(guess, game):
        if gu == ga:
            a += 1
        elif gu in game:
            b += 1
    return a, b

def play(run: int = 10, digits: int = 4, hint: bool = True) -> bool:
    '''play a guess-four-number game
    run    - maximum guess runs
    digits - digit numebrs of this game
    hint   - switch to show the answer of this game
    return - True for normal
             False for something wrong
    '''
    # validate arguments
    if type(run) != int or run <= 0:
        print('bad run')
        return False
    
    if type(digits) != int or digits < 1 or digits > 10:
        print('bad digits')
        return False
    
    # new game
    game = random.sample(range(10), k=digits)
    if hint:
        print(game)
    
    win = False
    for r in range(run):
        # user guess
        guess = input(f'[run {r+1:02}] 猜{digits}個數字: ')
        guess = _extract_guess(guess, digits)
        if not guess:
            print('bad input')
            continue
        
        # judge
        a, b = _judge(guess, game)
        print(f'{a}A {b}B')
        
        # check win
        if a == digits:
            win = True
            break
    
    print(f'恭喜第{r+1}輪獲勝' if win else '殘念')
    return True
````

````python
play(2, hint=True)
````
