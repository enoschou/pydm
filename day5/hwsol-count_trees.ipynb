## Solution

````python
import os
import glob


def count_trees(root: str = 'trees', filename: str = None) -> bool:
    '''output the count of tree species
    root     - root path of trees
    filename - file name for output in CSV; if not specified, output to the screen
    return   - True for success
               False for failure
    '''
    if not root or type(root) != str:
        return False

    # prepare the file handler
    file = None
    if filename and type(filename) == str:
        file = open(filename, 'w', encoding='utf-8')

    # aggregrate subsets
    subsets = []
    for s in os.listdir(root):
        if not s.startswith('.'):
            subsets.append(s)
    subsets.sort()

    # aggregrate categories
    cats = set()
    for s in subsets:
        for c in os.listdir(os.path.join(root, s)):
            if not c.startswith('.'):
                cats.add(c)
    cats = sorted(cats)

    # output column names
    print(','+','.join(cats), file=file)
    
    # count trees
    for s in subsets:
        row = [s]
        for c in cats:
            path = os.path.join(root, s, c, '*.[jJ][pP][gG]')
            len_s_c = len(glob.glob(path))
            row.append(str(len_s_c))
        print(','.join(row), file=file)  # output row

    # close file handler
    if file:
        file.close()

    return True

count_trees()
````
