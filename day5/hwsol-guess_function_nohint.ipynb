## Solution

```python
import random

def play(run: int = 10, digits: int = 4, hint: bool = True) -> bool:
    '''play a guess-four-number game
    run    - maximum guess runs
    digits - digit numebrs of this game
    hint   - switch to show the answer of this game
    return - True for normal
             False for something wrong
    '''
    if type(run) != int or run <= 0:
        print('bad run')
        return False
    
    if type(digits) != int or digits < 1 or digits > 10:
        print('bad digits')
        return False
    
    # new game
    game = random.sample(range(10), k=digits)
    if hint:
        print(game)
    
    win = False
    for r in range(run):
        # user guess
        guess = input(f'[run {r+1:02}] 猜{digits}個數字: ')
        guess = ''.join(guess.split())
        if not guess.isdigit() or len(set(guess)) != digits:
            print('bad input')
            continue
        guess = [int(n) for n in guess]
        
        # judge
        a = b = 0
        for ga, gu in zip(game, guess):
            if ga == gu:
                a += 1
            elif ga in guess:
                b += 1
        print(f'{a}A {b}B')
        
        # check win
        if a == digits:
            win = True
            break
    
    print(f'恭喜第{r+1}輪獲勝' if win else '殘念')
    return True
```

```python
play()
```
