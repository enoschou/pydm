## Solution

```python
import random

# 出題 (random)
game = random.sample(range(10), k=4)
print(game)
win = False

for i in range(2):
    # 猜數字 (input)
    done = False
    while not done:
        guess = input(f'[round {i+1:2d}] 猜4個數字:').replace(' ', '')
        if len(guess) == 4 and len(set(guess)) == 4 and guess.isdigit():
            done = True     
    guess = [int(g) for g in guess]
    print(guess)
    
    # 判斷 ?A?B
    a = b = 0
    for i, g in enumerate(guess):
        if game[i] == g:
            a += 1
        elif g in game:
            b += 1
    
    # 輸出結果
    print(f'{a}A, {b}B')
    if a == 4:
        win = True
        break
        
print(f"{'You Win!' if win else f'You Lose! the game is {game}'}")
```

```python
import random


def _new_game(digit: int = 4) -> list[int]:
    return random.sample(range(10), k=digit)

def _guess_input(i: int, digit: int = 4) -> list[int]:
    done = False
    while not done:
        guess = input(f'[round {i+1:2d}] 猜{digit}個數字:').replace(' ', '')
        if len(guess) == digit and len(set(guess)) == digit and guess.isdigit():
            done = True     
    return [int(g) for g in guess]

def _judge(game: list[int], guess: list[int]) -> tuple[int]:
    a = b = 0
    for i, g in enumerate(guess):
        if game[i] == g:
            a += 1
        elif g in game:
            b += 1
    return a, b

def main(digit: int = 4) -> None:
    if type(digit) != int or digit < 4 or digit > 5:
        digit = 4

    game = _new_game(digit)
    win = False

    for i in range(2):
        guess = _guess_input(i, digit)
        print(guess)
        
        a, b = _judge(game, guess)
        
        print(f'{a}A, {b}B')
        if a == 4:
            win = True
            break
            
    print(f"{'You Win!' if win else f'You Lose! the game is {game}'}")

main()
```
