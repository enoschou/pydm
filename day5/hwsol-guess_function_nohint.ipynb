## Solution

```python
from random import sample

def new():
    return sample(range(10), k=4)

def validate(guess):
    if not guess and type(guess) != str:
        return
        
    digits = ''.join(guess.split())
    if not digits.isdigit() or len(digits) != 4 or len(set(digits)) != 4:
        return

    return [int(d) for d in digits]

def judge(game, guess):
    a = b = 0
    for x, y in zip(game, guess):
        if x == y:
            a += 1
        elif x in guess:
            b += 1
    return a, b

def play(run=10):
    # init game
    game = new()
    print(f'new game {game}')
    
    count = 0
    win = False

    # start guessing
    while count < run and not win:
        # user input
        guess = input(f'[round {count}] 猜 4 個數字: ')
        guess = validate(guess)
        if not guess:
            print('bad input!')
            continue

        # judge result
        a, b = judge(game, guess)
        print(f'{a}A, {b}B')

        # check if user win
        if a == 4:
            win = True
            break
    
        count += 1
    
    # output final result
    print('you ' + ['lose', 'win'][win])

play()
```
