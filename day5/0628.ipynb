```python
def pace_calc(k, t):
    ts = t.split(':')
    s = int(ts[0]) * 60 + int(ts[1])
    pace = s / k
    #print(f'{int(pace // 60)}:{int(pace % 60)}')
    return int(pace // 60), int(pace % 60), 100
```

```python
def calc_lane(n: int = 1) -> float | str:
    '''calucate lane length
    n      - lane number, from 1 to ...
    return - the length of a lane, in meter
             error string if invalid n
    '''
    if type(n) != int:
        return 'bad input'
    if n <= 0 or n >= 10:
        return 'invalid lane'
    
    pi = 3.1416
    r = 36.5
    w = 1.22
    d1, d2 = 0.3, 0.2
    straight = 84.39
    
    i = n - 1
    
    d = d1 if i == 0 else d2
    lane = 2 * pi * (r + i * w + d) + 2 * straight
    
    return lane
```

```python
def egfr(creatinine: float, gender: str, age: float | int) -> float:
    value = 186 * (creatinine ** -1.154) * (age ** -0.203)
    return value if gender == 'm' else 0.742 * value

def stage(egfr: float) -> int:
    values = [100, 90, 60, 30, 15]
    for i, v in enumerate(values):
        if egfr >= v:
            return i
    return i + 1

m = input('性別(m/f): ')
a = float(input('年齡: '))
c = float(input('血清肌酸酐: '))

value = egfr(c, m, a)
print(value)
print(stage(value))
```

```python
# 隨機發牌給 player1, player2
from random import sample

SUITS = ['SPADE', 'HEART', 'DIAMOND', 'CLUB']
NUMBERS = ['2', 'A', 'K', 'Q', 'J', '10', '9', '8', '7', '6', '5', '4', '3']

cards = [(s, n) for s in SUITS for n in NUMBERS]

player1, player2 = sample(cards, k=2)
print(f'player1 {player1}, player2 {player2}')
```

```python
import random

SUITS = ['SPADE', 'HEART', 'DIAMOND', 'CLUB']
NUMBERS = ['2', 'A', 'K', 'Q', 'J', '10', '9', '8', '7', '6', '5', '4', '3']

def deal(n: int) -> list:
    # prepare 52 cards
    cards = [(s, _n) for s in SUITS for _n in NUMBERS]
    
    # shuffle
    random.shuffle(cards)
    
    # deal
    _cards = cards[:n]
    cards = cards[n:]
    return _cards
```

```python
deal(5)
```
