```python
import glob
import os

path_w = os.path.join('trees', 'train', '*')
cats = glob.glob(path_w)
print(cats)
```

```python
for c in cats:
    file_w = os.path.join(c, '*.jpg')
    files = glob.glob(file_w)
    name = os.path.basename(c)
    #name = c.split(os.path.sep)[-1]
    print(f'{name}: {len(files):3}')
```

```python
import glob
import os

path_w = os.path.join('trees', 'train')
cats = os.listdir(path_w)
print(cats)
```

```python
for c in cats:
    file_w = os.path.join(path_w, c, '*.jpg')
    files = glob.glob(file_w)
    name = os.path.basename(c)
    #name = c.split(os.path.sep)[-1]
    print(f'{name}: {len(files):3}')
```
