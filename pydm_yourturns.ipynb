```python
from random import choice

allow = ['剪刀', '石頭', '布']
user = input('請輸入剪刀、石頭、布')
if user in allow:
    computer = choice(allow)
    print(f'user {user} vs computer {computer}: ', end='')
    u_c = allow.index(user) - allow.index(computer)
    if not u_c:  # 0
        print('tie.')
    elif u_c in (-2, 1):
        print(f'user win!')
    else:  # in (-1, 2)
        print(f'user lose.')
else:
    print('輸入錯誤')
```

```python
# Please save as yourturn_5_2.py

import requests

class WeaG:
    '''Weather Grabber, to get recent weather information via open data provided by CWA.'''
    
    URL = 'https://opendata.cwa.gov.tw/api/v1/rest/datastore/'

    def __init__(self, key):
        '''Initialize the grabber with key.
        
        key - 授權碼 CWA open data
        '''
    
        self.AUTH = key
        self.params = {'Authorization': self.AUTH}
    
    def grab(self, location):
        '''Grab the temperature, humidity, and rainfall of specific location.
        
        location - weather site name, ex: 臺北
        return - observation time, temperature, humidity, and rainfall of location
                 None if location is invalid or other issues
        '''
        
        self.params['StationName'] = location
        if not (w := self._grab(self.URL + '/' + 'O-A0001-001', location)):
            w = self._grab(self.URL + '/' + 'O-A0003-001', location)
        r = self._grab(self.URL + '/' + 'O-A0002-001', location)
        if w:
            if r:
                w.update(r)
            return w
        if r:
            return r

    def _grab(self, url, station):
        r = requests.get(url, params=self.params)
        r.close()
        if r.status_code == 200:
            stations = r.json()['records']['Station']
            if w := self._extract(station, stations):
                return w
    
    def _extract(self, station, stations):
        for s in stations:
            if s['StationName'] == station:
                r = {}
                if 'ObsTime' in s and 'DateTime' in s['ObsTime']:
                    r ['O'] = s['ObsTime']['DateTime']
                if 'WeatherElement' in s:
                    if 'AirTemperature' in s['WeatherElement']:
                        r['T'] = float(s['WeatherElement']['AirTemperature'])
                    if 'RelativeHumidity' in s['WeatherElement']:
                        r['H'] = float(s['WeatherElement']['RelativeHumidity'])/ 100
                if 'RainfallElement' in s and 'Past24hr' in s['RainfallElement'] and 'Precipitation' in s['RainfallElement']['Past24hr']:
                    r['R'] = float(s['RainfallElement']['Past24hr']['Precipitation'])
                return r

if __name__ == '__main__':
    import argparse

    parser = argparse.ArgumentParser()
    parser.add_argument('location', help='氣象站名稱')
    parser.add_argument('-k', '--key', type=str, required=True, help='氣象局授權碼')
    args = parser.parse_args()
    
    w = WeaG(args.key)
    r = w.grab(args.location)
    if r:
        t = []
        if 'O' in r:
            t.append(f'觀測時間: {r["O"]}')
        if 'T' in r:
            t.append(f'溫度: {r["T"]:.1f}°C')
        if 'H' in r:
            t.append(f'濕度: {r["H"]:.0%}')
        if 'R' in r:
            t.append(f'雨量: {r["R"]:.1f}mm')
        print(f"{args.location} {', '.join(t)}")
    else:
        print(f'{args.location} 查無此站！')
```

```python
# Please save as yourturn_5_4.py

import requests
from bs4 import BeautifulSoup

class WeaG:
    
    def __init__(self):
        self._url = 'https://www.cwa.gov.tw/V8/C/W/Observe/MOD/24hr/TBD.html'
        self._load()
    
    def grab(self, location):
        rtn = {}
        if location in self._sites:
            url = self._url.replace('TBD', self._sites[location])
            r = requests.get(url)
            if r.status_code == 200:
                soup = BeautifulSoup(r.text, 'html.parser')
                try:
                    rtn['O'] = soup.find(headers="time").text
                    rtn['T'] = float(soup.find(class_="tem-C").text)
                    rtn['H'] = float(soup.find(headers="hum").text)/100
                    rtn['R'] = float(soup.find(headers="rain").text)
                except:
                    pass
            r.close()
        return rtn
    
    def _load(self):
        url = 'https://www.cwa.gov.tw/Data/js/Observe/OSM/C/STMap.json'
        j = requests.get(url)
        self._sites = {i['STname']: i['ID'] for i in j.json()} if j.status_code == 200 else {}
        j.close()

if __name__ == '__main__':
    import argparse

    parser = argparse.ArgumentParser()
    parser.add_argument('location', help='location weather to be grabbed')
    args = parser.parse_args()
    
    w = WeaG()
    r = w.grab(args.location)
    if r:
        print(f'{args.location} 觀測時間: {r["O"]} 溫度: {r["T"]:.1f}°C, 濕度: {r["H"]:.0%}, 雨量: {r["R"]:.1f}mm')
    else:
        print(f'{args.location} 查無此站！')
```
