```python
class CKDStager:    
    @staticmethod
    def stage(creatinine, gender, age):
        stages = [100, 90, 60, 30, 15, 0]
        egfr = CKDStager.egfr(creatinine, gender, age)
        for i, v in enumerate(stages):
            if egfr >= v:
                return i, egfr

    @staticmethod
    def egfr(creatinine, gender, age):
        val = 186 * (creatinine ** -1.154) * (age ** -0.203)
    
        if gender.lower() in ['female', 'f']:
            return val * 0.742
        return val
```

```python
gen = input('性別 (f / m)： ')
age = int(input('年齡： '))
cre = float(input('血清肌酸酐(mg/dL)： '))
q = input('計算 1.腎臟病分期 2.eGFR：')

d = {'m':'男性', 'f':'女性'}
if q == '1':
    s, e = CKDStager.stage(cre, gen, age)
    print(f'{d[gen]}，{age} 歲，血清肌酸酐 {cre} mg/dL ==> eGFR {e:.3f} mL/min/1.73m\xB2')
    print(f'腎臟病第 {s} 期')
elif q == '2':
    e = CKDStager.egfr(cre, gen, age)
    print(f'{d[gen]}，{age} 歲，血清肌酸酐 {cre} mg/dL ==> eGFR {e:.3f} mL/min/1.73m\xB2')
```
