```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.cwb.gov.tw/V8/C/W/Observe/MOD/24hr/46692.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
tr = soup.find('tr')
print(tr.find(headers="time"))
print(tr.find(headers="temp"))
print(tr.find(headers="hum"))
print(tr.find(headers="rain"))
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.cwb.gov.tw/V8/C/W/Observe/MOD/24hr/46692.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
tr = soup.find('tr')
print(f'時間： {tr.find(headers="time").text}')
print(f'溫度： {tr.find(headers="temp").find("span").text}')
print(f'濕度： {tr.find(headers="hum").text}')
print(f'雨量： {tr.find(headers="rain").text}')
```
