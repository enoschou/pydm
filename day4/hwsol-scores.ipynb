## Tasks
學生分數儲存方式如下 `scores = [11, 22, 33, 44, 55, 77, 99, 88, 66]`，請將分數排序，再逐行輸出及格最低分與不及格最高分。若沒有學生及格，則及格最低分改輸出為 `worst case`；若沒有學生不及格，則不及格最高分改輸出為 `best case`。
1. 以 `for` 完成
2. 以 `while` 完成

## Output
```text
# case 1
# scores = [11, 22, 33, 44, 55, 77, 99, 88, 66]
66
55

# case 2
# scores = [78, 63]
63
best case

# case 3
# scores = [13]
worst case
13
```

### 1. 以 `for` 完成

````python
scores = [11, 22, 33, 44, 55, 77, 99, 88, 66]
scores = [78, 63]
scores = [13]

# sort the scores
scores.sort()
print(scores)

# find the first pass
found = False
for i, s in enumerate(scores):
    if s >= 60:
        found = True
        break

# output
if found:  # found
    print(s)
    print(scores[i-1] if i else 'best case')
else:  # not found
    print(f'worst case')
    print(s)
````

### 2. 以 `while` 完成

````python
scores = [11, 22, 33, 44, 55, 77, 99, 88, 66]
scores = [78, 63]
scores = [13]

# sort the scores
scores.sort()
print(scores)

# find the first pass
i = 0
n = len(scores)
while i < n and scores[i] < 60:
    i += 1

print()

# output by found or not found
if i < n:  # found
    print(scores[i])
    print(scores[i-1] if i else 'best case')
else:  # not found
    print('worst case')
    print(scores[-1])

print()

# output by index
print('worst case' if i == n else scores[i])
print('best case' if i == 0 else scores[i-1])
````
