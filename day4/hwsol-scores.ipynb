## Solution

```python
# by for loop

scores = [11, 22, 33, 44, 55, 66, 99, 88, 77]
#scores = [78, 63]
#scores = [13]

scores.sort()
print(scores)

found = False
for i in range(len(scores)):
    if scores[i] > 60:
        print(scores[i])
        print('best case' if i == 0 else scores[i-1])
        found = True
        break

if not found:
    print(scores[i-1])
    print('worst case')
```

```python
# by while loop

scores = [11, 22, 33, 44, 55, 66, 99, 88, 77]
#scores = [78, 63]
#scores = [13]

scores.sort()
print(scores)

i = 0
n = len(scores)
while i < n and scores[i] < 60:
    i += 1

print('worst case' if i == n else scores[i])
print('best case' if i == 0 else scores[i-1])
```
