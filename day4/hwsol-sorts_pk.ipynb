## 實作兩種排序演算法，比較執行速度
實作細節如下：<br>
<ol>
    <li>運用 random 模組建立數量夠大 (> 2,000 筆) 的隨機整數 (範圍為 0 ~ 1000) list
    <li>實作兩種排序演算法 (bubble sort, selection sort, quick sort, ...) 針對上述 list 進行 in-place 排序
    <li>比較兩種排序演算法執行時間
    <li>盡可能優化上述程式以減少執行時間
</ol>

## Build random int list

```python
from time import time
from random import choices

nums = choices(range(1000), k=3000)
```

## Bubble Sort

```python
ns1 = nums.copy()
start = time()
i = 0
len_1 = len(ns1)-1
while i < len_1:  # control bubble sort loop times
    j = 0
    while j < len_1-i:  # largest number will rise up to the most right in each loop
        if ns1[j] > ns1[j+1]:
            ns1[j], ns1[j+1] = ns1[j+1], ns1[j]
        j += 1
    i += 1
print(f'{time()-start:.3f}s')
```

## Selection Sort

```python
ns2 = nums.copy()
start = time()
i = 0
len_ = len(ns2)
len_1 = len_-1
while i < len_1:  # control selection sort loop times
    small, j = i, i+1
    while j < len_:  # select and swap smallest number to the most left in ecah loop
        if ns2[small] > ns2[j]:
            small = j
        j += 1
    ns2[i], ns2[small] = ns2[small], ns2[i]
    i += 1
print(f'{time()-start:.3f}s')
```
