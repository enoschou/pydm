# Solution 三角形辨別

```python
# input
lens = [int(n) for n in input().split()]
```

```python
# input
lens = [int(n) for n in input().split()]

# bubble sort
sz = len(lens)
for i in range(sz-1):
    for j in range(sz-1-i):
        if lens[j] > lens[j+1]:
            lens[j], lens[j+1] = lens[j+1], lens[j]  # swap

# output
for n in lens:  # print(*lens)
    print(n, end=' ')
print()

# access and output
if lens[0] + lens[1] <= lens[2]:
    print('No')
else:
    x = lens[0]**2 + lens[1]**2
    y = lens[2]**2
    if x < y:
        print('Obtuse')
    elif x == y:
        print('Right')
    else:
        print('Acute')
```
